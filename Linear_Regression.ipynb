```python
# You need to import the pandas and seaborn library
import pandas as pan
import seaborn as sea
```

```python
# Read the data from the CSV file with read from pandas

data = pan.read_csv("Salary_dataset.csv")

# Configure the style
sea.set_theme(style="darkgrid")

# Plot the data points with seaborn
sea.jointplot(data=data, x="YearsExperience", y="Salary")

```

```python
# We're going to use this library and find out how it works on the Internet
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plot

# Separate the features (YearsExperience) from the target variable (Salary)
x = data[['YearsExperience']]
y = data['Salary']

# Create a linear regression model
regression = LinearRegression()

# Train the model
train = regression.fit(x, y)

# Predict the Salary values based on YearsExperience
predictions = regression.predict(x)



# Creating a comparative graph for actual data vs predicitions

# 1 - Concatenate the training data with the predicted values
train_data = pan.concat([x, y], axis=1)
train_data['Predicted Salary'] = predictions

# 2 - Create a linear regression plot with Seaborn's lmplot
sea.lmplot(data=train_data, x='YearsExperience', y='Salary', aspect=1.5, height=6)
sea.scatterplot(data=train_data, x='YearsExperience', y='Predicted Salary', color='red', label='Predicted')
plot.title('Linear Regression: Actual vs. Predicted (Training Data)')
plot.legend()
plot.show()
```
